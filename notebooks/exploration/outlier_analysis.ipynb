```python
import pandas as pd
import numpy as np
from scipy import stats
```

# County

```python
# Read data
df = pd.read_csv("../../data/final/county_data.csv")
df = df.set_index("fips")
df
```

```python
df.describe(percentiles=[0.01, 0.25, 0.5, 0.75, 0.99])
```

```python
# Typically you would use a z-score cutoff of 3 for outlier analysis but we will
# use 4 to capture only the most extreme
def get_outliers(column):
    df_cols = df[[column]].dropna()
    return df_cols[np.abs(stats.zscore(df_cols)) > 4]


for col in df.columns:
    print(get_outliers(col))
```
