# Lat/Long data

For these data files, we need to add the associated county FIPS code for each entry.

```python
files = [
    "oil_accidents.csv",
]
```

```python
def get_fips(lat, lon):
    # Encode parameters and construct request URL
    params = urllib.parse.urlencode({"latitude": lat, "longitude": lon, "format": "json"})
    url = "https://geo.fcc.gov/api/census/block/find?" + params

    # Get response from API
    response = requests.get(url)

    # Parse json in response
    data = response.json()

    # Print FIPS code
    return data["County"]["FIPS"]
```

```python
for f in files:
    df = pd.read_csv("../../data/processed/" + f)
    if "fips" not in df.columns:
        df["fips"] = df.apply(lambda row: get_fips(row["latitude"], row["longitude"]), axis=1)
    df.to_csv("../../data/final/" + f, index=False)
```
