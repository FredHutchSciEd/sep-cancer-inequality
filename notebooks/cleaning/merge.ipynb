# County data

Merge all of the county-level dataframes into a single file for analysis.

```python
# Start with the NCI dataset
df = pd.read_csv("../../data/processed/nci.csv")
df
```

```python
files = [
    "food_access.csv",
    "language_barriers.csv",
    "social_vulnerability_index.csv",
    "walkability.csv",
    "hospital_beds_total.csv",
    "air_quality.csv",
]
```

```python
for f in files:
    df2 = pd.read_csv("../../data/processed/" + f)
    df = df.set_index("fips").join(df2.set_index("fips"), how="outer").reset_index()
df
```

```python
df.columns
```

```python
# Grab the subset of columns that will be used for analysis
df = df[
    [
        "fips",
        "cancer_incidence_rate_per_100000",
        "cancer_mortality_rate_per_100000",
        "breast_cancer_incidence_rate_per_100000",
        "breast_cancer_mortality_rate_per_100000",
        "colorectal_cancer_incidence_rate_per_100000",
        "colorectal_cancer_mortality_rate_per_100000",
        "leukemia_cancer_incidence_rate_per_100000",
        "leukemia_cancer_mortality_rate_per_100000",
        "lung_cancer_incidence_rate_per_100000",
        "lung_cancer_mortality_rate_per_100000",
        "non_hodgkin_lymphoma_cancer_incidence_rate_per_100000",
        "non_hodgkin_lymphoma_cancer_mortality_rate_per_100000",
        "prostate_cancer_incidence_rate_per_100000",
        "prostate_cancer_mortality_rate_per_100000",
        "median_household_income",
        "over_65_percent",
        "uv_exposure",
        "low_income_low_access_share",
        "non_english_speaking",
        "population_in_poverty_percent",
        "population_over_25_no_high_school_diploma_percent",
        "population_uninsured_percent",
        "population_minority_percent",
        "walkability_index",
        "hospital_beds_per_100000",
        "median_aqi",
    ]
]
```

```python
# US State FIPS codes go from 01 to 56 so anything greater than that is a territory.
df = df[df["fips"] < 57000]
df
```

```python
def merge_counties(df, old, new):
    df.loc[df["fips"] == old, "fips"] = new
    df = df.groupby("fips").max().reset_index()
    return df
```

```python
# Shannon County, SD (fips 46113) was renamed to Oglala Lakota County (fips 46102)
df = merge_counties(df, 46113, 46102)

# Wade Hampton Census Area, AK (fips 02270) was renamed to Kusilvak Census Area (fips 02158)
df = merge_counties(df, 2270, 2158)

# City of Bedford, VA (fips 51515) and Beford County (fips 51917) were merged to Bedford County (fips 51019)
df = merge_counties(df, 51515, 51019)
df = merge_counties(df, 51917, 51019)

# Wrangell-Petersburg (fips 02280) was split into Wrangell City (fips 02275) and Petersburg Census Area (fips 02195)
df.loc[df["fips"] == 2280, "fips"] = 2275
new = pd.DataFrame(df.loc[df["fips"] == 2275])
new.loc[new["fips"] == 2275, "fips"] = 2195
df = pd.concat([df, new])
df = df.groupby("fips").max().reset_index()
```

```python
# Write out the merged data
df.to_csv("../../data/final/county_data.csv", index=False)
```

# Lat/Long data

For these data files, we need to add the associated county FIPS code for each entry.

```python
files = [
    "oil_accidents.csv",
]
```

```python
def get_fips(lat, lon):
    # Encode parameters and construct request URL
    params = urllib.parse.urlencode({"latitude": lat, "longitude": lon, "format": "json"})
    url = "https://geo.fcc.gov/api/census/block/find?" + params

    # Get response from API
    response = requests.get(url)

    # Parse json in response
    data = response.json()

    # Print FIPS code
    return data["County"]["FIPS"]
```

```python
for f in files:
    df = pd.read_csv("../../data/processed/" + f)
    if "fips" not in df.columns:
        df["fips"] = df.apply(
            lambda row: get_fips(row["latitude"], row["longitude"]), axis=1
        )
    df = df[df["fips"].notna()]
    df.to_csv("../../data/final/" + f, index=False)
```
