# County data

Merge all of the county-level dataframes into a single file for analysis.

```python
# Start with the NCI dataset
df = pd.read_csv("../../data/processed/nci.csv")
df
```

```python
files = [
    "food_access.csv",
    "language_barriers.csv",
    "social_vulnerability_index.csv",
    "walkability.csv"
]
```

```python
for f in files:
    df2 = pd.read_csv("../../data/processed/" + f)
    df = df.set_index("fips").join(df2.set_index("fips"), how="outer").reset_index()
df
```

```python
df.columns
```

```python
# Grab the subset of columns that will be used for analysis
df = df[
    [
        "fips",
        "cancer_incidence_rate_per_100000",
        "cancer_mortality_rate_per_100000",
        "breast_cancer_incidence_rate_per_100000",
        "breast_cancer_mortality_rate_per_100000",
        "colorectal_cancer_incidence_rate_per_100000",
        "colorectal_cancer_mortality_rate_per_100000",
        "lung_cancer_incidence_rate_per_100000",
        "lung_cancer_mortality_rate_per_100000",
        "prostate_cancer_incidence_rate_per_100000",
        "prostate_cancer_mortality_rate_per_100000",
        "median_household_income",
        "language_isolation_percent",
        "over_65_percent",
        "below_poverty_percent",
        "uninsured_percent",
        "uv_exposure",
        "low_income_low_access_share",
        "non_english_speaking",
        "population_in_poverty_percent",
        "population_over_16_unemployed_percent",
        "population_over_25_no_high_school_diploma_percent",
        "population_over_65_percent",
        "population_under_17_percent",
        "population_uninsured_percent",
        "population_with_disability_percent",
        "population_minority_percent",
        "walkability_index",
    ]
]
```

```python
# Write out the merged data
df.to_csv("../../data/final/county_data.csv", index=False)
```
