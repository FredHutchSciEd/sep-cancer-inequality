```python
import pandas as pd
```

```python
# Read in raw data
df = pd.read_csv("../../data/raw/database.csv")
df
```

```python
# Only keep a subset of columns
df = df[
    [
        "Report Number",
        "Accident Latitude",
        "Accident Longitude",
        "Unintentional Release (Barrels)",
        "Intentional Release (Barrels)",
    ]
]
df
```

```python
# Group by lat/long to get total spills in each location
df = (
    df.groupby(["Accident Latitude", "Accident Longitude"])
    .agg(
        {
            "Unintentional Release (Barrels)": "sum",
            "Intentional Release (Barrels)": "sum",
            "Report Number": "count",
        }
    )
    .reset_index()
)
df
```

```python
# Rename columns
df["oil_barrels_spilled"] = (
    df["Unintentional Release (Barrels)"] + df["Intentional Release (Barrels)"]
)
df = df.rename(
    columns={
        "Accident Latitude": "latitude",
        "Accident Longitude": "longitude",
        "Report Number": "spill_count",
    }
)
df = df[["latitude", "longitude", "oil_barrels_spilled", "spill_count"]]
df
```

```python
# Write out the cleaned data
df.to_csv("../../data/processed/oil_accidents.csv", index=False)
```
