```python
import json
import pandas as pd
```

```python
# Read in the raw data
geo = json.load(open("../../data/raw/fullDownload.geojson"))
geo
```

```python
# Examine how sample data looks
geo['features'][1]
```

```python
# Turn into a dataframe and extract desired columns
df = pd.DataFrame(
    list(
        map(
            lambda f: [
                f["properties"]["state"],
                f["properties"]["city"],
                f["properties"]["holc_grade"],
                f["properties"]["holc_id"] if f["properties"]["name"] == None else f["properties"]["name"],
                None if f["geometry"] == None else f["geometry"]["coordinates"]
            ],
            geo["features"],
        )
    ),
    columns=["state", "city", "holc_grade", "name", "geometry"],
)
df
```

```python
# Write out the cleaned data
df.to_csv("../../data/processed/redlining.csv", index=False)
```
