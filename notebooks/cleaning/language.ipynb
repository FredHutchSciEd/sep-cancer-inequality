```python
import pandas as pd
import numpy as np
```

```python
# Read in raw data
df = pd.read_csv("../../data/raw/ACSDP5Y2020.DP02-Data.csv", header=1)
df
```

```python
# Simplify to the desired columns
df = df.rename(
    columns={
        "Percent!!LANGUAGE SPOKEN AT HOME!!Population 5 years and over!!Language other than English!!Speak English less than 'very well'": "non_english_speaking"
    }
)
df = df[["Geography", "non_english_speaking"]]
df
```

```python
# Convert "Geography" to fips
df["fips"] = df["Geography"].str.replace("0500000US", "").astype(int)
df
```

```python
# Replace N with NaN
df = df.replace("N", np.nan)
df["non_english_speaking"] = df["non_english_speaking"].astype(float)
df
```

```python
# Simplify to desired columns
df = df[["fips", "non_english_speaking"]]
df
```

```python
# Write out the cleaned data
df.to_csv("../../data/processed/language_barriers.csv", index=False)
```
