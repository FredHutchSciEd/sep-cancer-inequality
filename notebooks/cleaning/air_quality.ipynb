```python
import pandas as pd
```

```python
# Read in data used to generate a FIPS code map
fips = pd.read_csv(
    "../../data/raw/Social_Vulnerability_Index_2020_US_COUNTY.csv", engine="python"
)
fips = fips[["STATE", "COUNTY", "FIPS"]]
fips = fips.rename(columns={"STATE": "State", "COUNTY": "County", "FIPS": "fips"})
fips
```

```python
# Read in raw data
df = pd.read_csv(
    "../../data/raw/annual_aqi_by_county_2010.csv", engine="python"
)
df
```

```python
# Remove rows outside the US
df = df[(df["State"] != "Country Of Mexico") & (df["State"] != "Puerto Rico") & (df["State"] != "Virgin Islands")]

# Fix county names
df.loc[(df["State"] == "Virginia") & (df["County"] == "Charles"), "County"] = "Charles City"
df.loc[(df["State"] == "New Mexico") & (df["County"] == "Dona Ana"), "County"] = "DoÃ±a Ana"
df.loc[(df["State"] == "Minnesota") & (df["County"] == "Saint Louis"), "County"] = "St. Louis"
df.loc[(df["State"] == "Missouri") & (df["County"] == "Saint Louis"), "County"] = "St. Louis"
df.loc[(df["State"] == "Missouri") & (df["County"] == "Saint Charles"), "County"] = "St. Charles"
df.loc[(df["State"] == "Missouri") & (df["County"] == "Sainte Genevieve"), "County"] = "Ste. Genevieve"
df.loc[(df["State"] == "Maryland") & (df["County"] == "Baltimore (City)"), "County"] = "Baltimore City"
df.loc[(df["State"] == "Illinois") & (df["County"] == "Saint Clair"), "County"] = "St. Clair"
df.loc[(df["State"] == "Illinois") & (df["County"] == "La Salle"), "County"] = "LaSalle"
df.loc[(df["State"] == "District Of Columbia") & (df["County"] == "District of Columbia"), "State"] = "District of Columbia"
df.loc[(df["State"] == "Alaska") & (df["County"] == "Kenai Peninsula "), "County"] = "Kenai Peninsula"
df.loc[(df["State"] == "Alaska") & (df["County"] == "Juneau "), "County"] = "Juneau City and Borough"
df.loc[(df["State"] == "Alaska") & (df["County"] == "Aleutians East "), "County"] = "Aleutians East"
df.loc[(df["State"] == "Alaska") & (df["County"] == "Anchorage "), "County"] = "Anchorage"
df.loc[(df["State"] == "Alaska") & (df["County"] == "Denali "), "County"] = "Denali"
df.loc[(df["State"] == "Alaska") & (df["County"] == "Fairbanks North Star "), "County"] = "Fairbanks North Star"
df.loc[(df["State"] == "Alaska") & (df["County"] == "Matanuska-Susitna "), "County"] = "Matanuska-Susitna"
df.loc[(df["State"] == "Alaska") & (df["County"] == "Yukon-Koyukuk "), "County"] = "Yukon-Koyukuk"

df
```

```python
df = df.set_index(["State", "County"]).join(fips.set_index(["State", "County"])).reset_index()
df
```

```python
df["unhealthy_air_days"] = df["Unhealthy for Sensitive Groups Days"] + df["Unhealthy Days"] + df["Very Unhealthy Days"] + df["Hazardous Days"]
df = df.rename(columns = {"Median AQI": "median_aqi"})
df
```

```python
df = df[["fips", "unhealthy_air_days", "median_aqi"]]
df
```

```python
# Write out the cleaned data
df.to_csv("../../data/processed/air_quality.csv", index=False)
```
